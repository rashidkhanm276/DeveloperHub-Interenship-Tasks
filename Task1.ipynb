Step 1: Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")

# pandas (pd) → used to work with data (tables)
# matplotlib (plt) → used to create charts
# seaborn (sns) → used to make better-looking charts
```

Step 2: Load Dataset

```python
df = pd.read_csv(r"G:\Data Science\DS 27\Data\Iris.csv")
df.head()

# load_dataset("iris") → loads the dataset
# df → variable where data is stored
# head() → shows first 5 rows of the dataset
```

Step 3: Explore Dataset Structure

```python
# Shape of dataset (rows, columns)
print("Shape:", df.shape)

# Column names
print("Columns:", df.columns)

# First few rows
print(df.head())

# Data types
print(df.info())

# Summary statistics
print(df.describe())
```

Step 4: Scatter Plot (Relationship Analysis)

```python
# Scatter plot: Sepal Length vs Sepal Width
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="sepal.length", y="sepal.width", hue="variety")

plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()

# Scatter plot shows relationship between two columns
# Here:
# X-axis → sepal_length
# Y-axis → sepal_width
# hue → color by species

# Each dot = one flower
```

Step 5: Histogram (Data Distribution)

```python
# Histogram for sepal length
plt.figure(figsize=(6,4))
sns.histplot(df["sepal.length"], kde=True)

plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()


# Histogram shows data distribution

# Example:
# How many flowers have small length
# How many have large length

# Why we use it?
# To understand:
# Data spread
# Shape (normal or not)
```

Step 6: Box Plot (Outliers Detection)

```python
# Box plot for petal length
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="variety", y="petal.length")

plt.title("Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.show()


# Box plot shows:

# Minimum value
# Maximum value
# Median (middle)
# Outliers

# It compares values for each species
```

Step 7: Pairplot

```python
# Pairplot (multiple relationships)
sns.pairplot(df, hue="variety")
plt.show()


# Shows multiple scatter plots together
# Compares all columns with each other

# Why we use it?
   # To quickly understand:
   # Relationships
   # Patterns
   # Class separation
```
